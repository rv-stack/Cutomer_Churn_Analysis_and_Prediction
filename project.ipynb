```python
df['Churn'] = df['Churn'].replace({'Yes': 1, 'No': 0})
df['Churn'] = df['Churn'].astype(int)
```

```python
object_columns = df.select_dtypes(include='object').columns
```

```python
df.head(
)
```

```python
from sklearn.preprocessing import LabelEncoder

encoders = {}

for column in object_columns:
    label_encoder = LabelEncoder()
    df[column] = label_encoder.fit_transform(df[column])
    encoders[column] = label_encoder
```

```python
encoders
```

```python
import pickle
with open('encoder.pkl', 'wb') as f:
    pickle.dump(encoders, f)
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler

numric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
scaler = StandardScaler()

df[numric_cols] = scaler.fit_transform(df[numric_cols])
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
df.describe()
```

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.22, random_state=42)
```

```python
y_train.value_counts(
)
```

```python
smote = SMOTE(random_state=42)
```

```python
X_train_somte, y_train_somte = smote.fit_resample(X_train, y_train)
```

```python
y_train_somte.value_counts()
```

```python
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42)
}
```

```python
hyperparameters = {
    "Random Forest": {
        'n_estimators': [50, 100, 200],
        'max_depth': [5, 10, None],
    },
    "XGBoost": {
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 7],
    }
}
```

```python
best_models = {}
for model_name , model in models.items():
    print(f'Turning {model_name}......')
    grid = GridSearchCV(estimator=model, param_grid=hyperparameters[model_name], cv=5, scoring='accuracy')
    grid.fit(X_train_somte, y_train_somte)
    best_models[model_name] = grid.best_estimator_
    print(f"Best Parametors for {model_name} : {grid.best_params_}")
    print(f"Best Acurracy for {model_name} : {grid.best_score_: .2f}\n")
```

```python
best_models
```

```python
with open('best_model.pkl', 'wb') as f:
    pickle.dump(best_models['Random Forest'], f)
```

```python
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

```python
y_test_pred = best_models['Random Forest'].predict(X_test)
y_test_prob = best_models['Random Forest'].predict_proba(X_test)[:,1]
```

```python
y_test_pred
```

```python
y_test_prob
```

```python
print(f'Accuracy :', accuracy_score(y_test, y_test_pred))
print(f'Roc - Auc score :', roc_auc_score(y_test, y_test_pred))
print(f'Confusion Matrix : \n', confusion_matrix(y_test, y_test_pred))
print(f'Classification Report : \n', classification_report(y_test, y_test_pred))
```

```python
with open('best_model.pkl', 'rb') as f:
    loded_model = pickle.load(f)
with open('encoder.pkl', 'rb') as f:
    encoders = pickle.load(f)
with open('scaler.pkl', 'rb') as f:
    scaler_data = pickle.load(f)
```

```python
def make_prediction(input_data):
    input_df = pd.DataFrame([input_data])
    for col, encoder in encoders.items():
        input_df[col] = encoder.transform(input_df[col])
    numric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
    input_df[numric_cols] = scaler_data.transform(input_df[numric_cols])
    prediction = loded_model.predict(input_df)[0]
    probability = loded_model.predict_proba(input_df)[0, 1]
    return "Churn" == 1 if prediction else "No Chrun", probability
```
